```python
#MInMaxScaler 
#formula :  x_new=(x_old-min)/(max-min)
```

```python
#suppose we create a array with the help of numpy (1 D array)
a=np.array([10,20,30,40,50])
b=np.array([100,200,300,400,500])
```

```python
print(a)
print(b)
```

```python
#create a user defined function 
def minmax(ar): #passing arguments/parameters
    ar_new=(ar-ar.min())/(ar.max()-ar.min()) #0  0.25  0.50   0.75 1
    return ar_new
```

```python
#main program
print(minmax(a))
```

```python
print(minmax(b))
```
